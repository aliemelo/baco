# Análise descritiva dos dados

```python
import pandas as pd

df_wine = pd.read_csv("wine-reviews/winemag-data-130k-v2.csv", index_col=0)

display(df_wine)
```

```python
# Describe the type of value of each column
df_wine.dtypes
```

```python
# Get the number of lines and how they are incremented
df_wine.index
```

```python
# A simple descriptive statistical analysis
df_wine.describe()
```

```python
# remover rows repetidas
df_wine_no_duplicates = df_wine.drop_duplicates()
print("Numero de linhas totais:",len(df_wine))
print("Numero de linhas apos retirar linhas duplicadas:",len(df_wine_no_duplicates))
```

```python
# checar quantos valores em branco dentro de cada coluna
print(df_wine_no_duplicates.isnull().sum())
```

```python
# extrair a safra da coluna 'title' e adicionar essa informacao em outra coluna

df_wine_no_duplicates['vintage'] = df_wine_no_duplicates.title.str.extract(r'([0-9]+)', expand=True)
df_wine_no_duplicates
```

```python
# checar os tipos novamente
df_wine_no_duplicates.dtypes
```

```python
import numpy as np

# substituir NaN na coluna vintage por zero
df_wine_no_duplicates["vintage"] = df_wine_no_duplicates["vintage"].replace(np.nan, 0, regex=True)

# converter o tipo da coluna vintage para int
df_wine_no_duplicates["vintage"] = df_wine_no_duplicates["vintage"].astype(dtype=np.int64)

# checar tipos
df_wine_no_duplicates.dtypes
```

```python
# checar quantos valores em branco dentro de cada coluna
print(df_wine_no_duplicates.isnull().sum())
```

```python
# salvar dataframe num arquivo

df_wine_no_duplicates.to_csv('wine_filtered.csv')
```
